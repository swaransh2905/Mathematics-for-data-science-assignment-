## Objectives:

# - Find where temperature changes fastest (Calculus)

```python
import numpy as np

#Coefficient Matrix A

A = np.array([
    [4, -1, -1, 0],
    [-1, 4, 0, -1],
    [-1, 0, 4, -1],
    [0, -1, -1, 4]
], dtype=float)

#vector b

b = np.array([100, 100, 0, 0], dtype=float)

#solving the linear equation

x = np.linalg.solve(A, b)

print("heat flow approximations at four grid points:")
for i, val in enumerate(x, 1):
    print(f" x{i} = {val:.2f} units")
```

```python
#fastest temperature change points

import matplotlib.pyplot as plt
import numpy as np

#the temperature function

x_vals= np.linspace(0, 10, 200)
T_vals = 100*np.sin(np.pi*x_vals/10)
dT_vals= 100 * (np.pi / 10) * np.cos(np.pi * x_vals / 10)

#maximum rate change

max_rate_points=[0,10]

print(f"temperature changes fast at x={max_rate_points}")
print("\n\n")

#plot

plt.figure(figsize=(8,5))
plt.plot(x_vals, T_vals, label="Temperature T(x)")
plt.plot(x_vals, dT_vals, label="Rate of Change dT/dx")
plt.legend()
plt.xlabel("x (cm)")
plt.ylabel("Temperature / Rate")
plt.title("Temperature Distribution and Rate of Change Along Plate Edge")
plt.grid(True)
plt.show()
```

```python
#Probability

from scipy.stats import norm

mu, sigma = 0, 2
prob = 1 - norm.cdf(3, mu, sigma)

print(f"P(random error >3Â°C) = {prob:.4f}")
print(f"{prob*100:.2f}% chance of error exceeding 3Â°C")
```

```python
#Hypothesis Test â True Mean = 100Â°C

from scipy import stats
from scipy.stats import ttest_1samp
import numpy as np

data = np.array([99, 100, 98, 101, 97, 99, 100, 98])

mean_sample = np.mean(data)
std_dev = np.std(data, ddof=1)

#t test

t_stat, p_value = ttest_1samp(data, 100)

print(f"sample mean= {mean_sample:.2f} Â°C")
print(f"sample std dev = {std_dev:.2f}")
print(f"t-statistic = {t_stat:.2f}")
print(f"p-value = {p_value:.4f}")

print("\n")

if p_value < 0.05:
    print("Reject the null hypothesis. The true mean is not equal to 100Â°C.")

else:
    print("Fail to reject the null hypothesis. The true mean is equal to 100Â°C.")


```
